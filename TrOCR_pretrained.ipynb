```python
import cv2
import numpy as np

def preprocess_image(image_path, output_path, target_width=768):
    # Load the image
    image = cv2.imread(image_path)

    # Denoise
    image_denoised = cv2.fastNlMeansDenoising(
        src=image,
        dst=None,
        h=25,
        templateWindowSize=7,
        searchWindowSize=21
    )

    # Convert to grayscale
    image_gray = cv2.cvtColor(image_denoised, cv2.COLOR_BGR2GRAY)

    # Apply CLAHE (Contrast Limited Adaptive Histogram Equalization)
    clahe = cv2.createCLAHE(clipLimit=1, tileGridSize=(4, 4))
    image_clahe = clahe.apply(image_gray)

    # Dilate (which is equivalent to eroding the background)
    kernel = np.ones((3, 3), np.uint8)
    image_dilated = cv2.erode(image_clahe, kernel, iterations=4)

    # Resize
    height, width = image_dilated.shape
    new_height = int(height * (target_width / width))
    image_resized = cv2.resize(image_dilated, (target_width, new_height))

    # Save the processed image
    cv2.imwrite(output_path, image_resized)
```

```python
import os
from tqdm import tqdm  # Add this import

# Process all images with progress tracking
input_dir = '/content/data/pages/'
output_dir = '/content/data/preprocessed/'
os.makedirs(output_dir, exist_ok=True)

# Add a progress bar using tqdm
image_files = [f for f in os.listdir(input_dir) if f.endswith('.jpg')]
for filename in tqdm(image_files, desc='Processing Images'):
    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, filename.replace('.jpg', '.png'))
    preprocess_image(input_path, output_path)

print("Preprocessing complete.")
```

Next: extract text line coordinates from XML files.
Contain information about layout of the document, including where each line of text is located.

-It opens and parses the XML file.

-It searches for all "TextLine" elements in the XML.

-For each TextLine, it looks for a "Coords" element.

-If it finds coordinates, it extracts the "points" attribute, which contains the coordinates of the text line.

-It stores these coordinates and returns a list of all text lines found.

```python
def parse_page_xml(xml_path):
    tree = ET.parse(xml_path)  # Parse the XML file
    root = tree.getroot()  # Get the root element of the XML
    lines = []
    # Find all TextLine elements in the XML
    for text_line in root.findall(".//TextLine"):
        coords = text_line.find("Coords")  # Find the Coords element for each TextLine
        if coords is not None:
            # Extract the points attribute and split it into a list
            points = coords.get("points").split()
            lines.append(points)
    return lines
```

Next function takes the extracted line coordinates and formats them into a JSON structure that the doc-ufcn model expects. JSON is a common format for storing and transmitting data.
The function creates a **dictionary** (which will be converted to JSON) that includes:


*   The width and height of the image

*   A list of all text lines, where each line is represented by its coordinates




```python
# Function to create JSON label in the format expected by doc-ufcn
def create_json_label(lines, image_width, image_height):
    label = {
        "img_w": image_width,
        "img_h": image_height,
        "lines": [
            {
                "points": line,
                "text": "",  # We don't need the text for line detection
                "class": "text_line"
            } for line in lines
        ]
    }
    return label
```
